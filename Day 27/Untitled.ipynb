## Missing data

```python
import pandas as pd
df3 =pd.read_csv("propData_c.csv")
```

```python
df = df3
df
```

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=1, how="any")
```

```python
# mean
avg = df["NUM_BEDROOMS"].mean(axis=0)
avg
```

```python
mod = df["NUM_BEDROOMS"].mode()[0]
mod
```

```python
med = df["NUM_BEDROOMS"].median()
med
```

```python
df["NUM_BEDROOMS"]= df["NUM_BEDROOMS"].fillna(avg)
df
```

```python
df["ST_NUM"].fillna(125, inplace=True)
df
```

```python
df["PID"] = df["PID"].fillna(method = 'bfill') # fill  with next value  / ffill fill with back value
df
```
