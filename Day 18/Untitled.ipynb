```python
import pandas as pd
```

```python
df = pd.read_csv("emp.csv")
```

```python
df
```

Agregate 

```python
df.agg(TotalEmp =("empno",'count'), AvgSal = ('sal','mean'), SumSal = ('sal','sum'))
```

```python
df.agg(TotalEmp =("empno",'count'), AvgSal = ('sal','mean'), SumSal = ('sal','sum')).T
```

```python
df2 =  df.groupby('deptno') .agg(TotalEmp =("empno",'count'), AvgSal = ('sal','mean'), SumSal = ('sal','sum')).reset_index()
df2
```

```python
df2 =  df.groupby('job') .agg(TotalEmp =("empno",'count'), AvgSal = ('sal','mean'), SumSal = ('sal','sum')).reset_index()
df2
```

```python
df2 =  df.groupby(['deptno','job']) .agg(TotalEmp =("empno",'count'), AvgSal = ('sal','mean'), SumSal = ('sal','sum')).reset_index()
df2
```

```python
e = df
d = pd.read_csv("dept.csv")
print(e.columns)
print(d.columns)
```

```python
pd.merge(e,d)
```

```python
 pd.merge(e,d, on='deptno', how='inner')
```

merge two columns with difrent names

```python
 pd.merge(e,d,  left_on='deptno', right_on='deptno' ,how='inner')
```

```python
 pd.merge(e,d, on='deptno', how='right') # 'left'; 'outer'
```

```python
pd.merge(e,d, on='deptno', how='outer', indicator=True)
```
